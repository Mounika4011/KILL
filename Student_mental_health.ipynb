```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

#reeading a dataset
dataset = pd.read_csv('/content/Student Mental health.csv')
dataset.head()
```

```python
print(dataset.shape)
dataset.info()
```

```python
print(dataset.isna().sum())
dataset.describe()
```

```python
copy_dataset = dataset.copy()
```

```python
print(copy_dataset.columns)
```

```python
copy_dataset.columns = ['Timestamp', 'Gender', 'Age', 'Course',
       'Year of Study', 'CGPA', 'Marital status',
       'Depression', 'Anxiety',
       'Panic attack',
       'Treatment']
copy_dataset
```

```python
copy_dataset['Year of Study'].unique()
```

```python
copy_dataset['Year of Study'].replace({'year 1': 'year 1', 'year 2': 'year 2', 'Year 1': 'year 1', 'year 3': 'year 3', 'year 4': 'year 4', 'Year 2': 'year 2',
       'Year 3': 'year 3'}, inplace = True)
copy_dataset['Year of Study'].unique()
```

```python
copy_dataset['Course'].unique()
```

```python
copy_dataset['Course'].replace({'engin': 'Engineering' , 'Engine':'Engineering' , 'Islamic education':'Islamic Education' , 'Pendidikan islam':'Islamic Education' , 'Pendidikan Islam ':'Islamic Education', 'BIT':'IT', 'psychology':'Psychology', 'koe': 'KOE','Koe': 'KOE', 'Kirkhs': 'KIRKHS', 'BENL': 'Benl', 'Fiqh fatwa ': 'Fiqh', 'Laws': 'Law', 'Econs': 'Economics'} , inplace = True)
copy_dataset['Course'].unique()
```

```python
copy_dataset['CGPA'].unique()
```

```python
copy_dataset['CGPA'] = copy_dataset['CGPA'].str.strip()
copy_dataset['CGPA'].unique()
```

```python
print(copy_dataset.isna().sum())
```

```python
copy_dataset['Age'].median()
```

```python
copy_dataset['Age'].mean()
```

```python
copy_dataset['Age'].fillna(copy_dataset['Age'].mean(), inplace = True)
```

```python
print(copy_dataset.isna().sum())
```

```python
copy_dataset.info()
```

```python
copy_dataset['Age']= copy_dataset['Age'].astype(int)
```

```python
copy_dataset.info()
```

```python
copy_dataset['Depression'].unique()
```

```python
copy_dataset['Depression'] = copy_dataset['Depression'].replace({'Yes': 'Depressed', 'No': 'Not Depressed'})
```

```python
copy_dataset['Depression'].unique()
```

```python
copy_dataset['Panic attack'].unique()
```

```python
copy_dataset['Panic attack'] = copy_dataset['Panic attack'].replace({'Yes': 'Panic Attack', 'No': 'Not Panic Attack'})
```

```python
copy_dataset['Panic attack'].unique()
```

```python
copy_dataset['Anxiety'].unique()
```

```python
copy_dataset['Anxiety'] = copy_dataset['Anxiety'].replace({'Yes': 'Anxiety Existing', 'No': 'Anxiety None'})
```

```python
copy_dataset['Anxiety'].unique()
```

```python
copy_dataset['Treatment'].unique()
```

```python
copy_dataset['Treatment'] = copy_dataset['Treatment'].replace({'Yes': 'Treatment', 'No': 'No Treatment'})
```

```python
copy_dataset['Treatment'].unique()
```

```python
copy_dataset['Marital status'].unique()
```

```python
copy_dataset['Marital status'] = copy_dataset['Marital status'].replace({'Yes': 'Married', 'No': 'Not Married'})
```

```python
copy_dataset['Marital status'].unique()
```

```python
colors = {'Panic Attack': '#C0D2EB', 'Not Panic Attack': '#6EA0E6'}
plt.figure(figsize=(8, 5))
sns.displot(data=df_depressed, x='Age', hue='Panic attack', palette=colors, kind='kde', fill=True)
plt.title('Distribution of Age with Panic Attack', fontsize=12)
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

```python
colors = {'Female': '#F5B5D2', 'Male': '#7FBADB'}
plt.figure(figsize=(10, 8))
sns.swarmplot(data=df_depressed, x='Anxiety', y='CGPA', hue='Gender', palette=colors)
plt.title('Swarm Plot of Anxiety by CGPA and Gender')
plt.xlabel('Anxiety')
plt.ylabel('CGPA')
plt.show()
```

```python
colors = {'Female': '#F5B5D2', 'Male': '#7FBADB'}
plt.figure(figsize=(10, 8))
sns.swarmplot(data=df_depressed, x='Anxiety', y='Year of Study', hue='Gender', palette= colors)
plt.title('Swarm Plot of Anxiety by year of study and Gender')
plt.xlabel('Anxiety')
plt.ylabel('Year of Study')
plt.show()
```
